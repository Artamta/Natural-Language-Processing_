# NLTK : Basics

```python
import nltk
nltk.data.path.append('/Users/ayush/Desktop/NLP/Self_study')

from nltk.corpus import reuters
```

```python
train_docs = [doc for doc in reuters.fileids() if doc.startswith("training/")]
test_docs = [doc for doc in reuters.fileids() if doc.startswith("test/")]

print(len(train_docs), len(test_docs))
```

```python
doc = train_docs[0]

print("File ID:", doc)
print("Labels:", reuters.categories(doc))
print("\nText preview:\n")
print(reuters.raw(doc)[:500])
```

```python
# Mking it singal catagary

# Prepare training data
X_train = []
y_train = []

for doc in train_docs:
    X_train.append(reuters.raw(doc))
    y_train.append(reuters.categories(doc)[0])  # take first label only

# Prepare test data
X_test = []
y_test = []

for doc in test_docs:
    X_test.append(reuters.raw(doc))
    y_test.append(reuters.categories(doc)[0])

print(len(X_train), len(y_train))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

print(X_train_vec.shape)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_vec, y_train)

accuracy = model.score(X_test_vec, y_test)
print("Accuracy:", accuracy)
```
